####  Dealing with missing values

```python
dataset.isnull()
```

```python
dataset.isnull().sum()
```

Using List Comprehension to create a list with the feature names

```python
[features for features in dataset.columns]
```

Getting the features which had null values in it..

```python
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>=1]
```

```python
features_with_na
```

% of Nan values present in the feature

```python
for features in features_with_na:
    print(features, np.round(dataset[features].isnull().mean()*100,5),'%')
```

#### Relationship between independent and dependent features

```python
import seaborn as sns
sns.boxplot(dataset['SalePrice'])
```

The dependent variable, SalePrice has alot of outliers.

If the independent features have missig values, it is always better to understand the relationship between the independent and dependent feature before handling the missing values in independent features.

```python
# Making a copy of the dataset

data= dataset.copy()
```

```python
for features in features_with_na:
    
    # creating a variable which records '1' when observation is missing and '0',otherwise.
    data[features]=np.where(data[features].isnull(),1,0)
    # calculating the mean of 'SalePrice' where data is missing or present
    data.groupby(features)['SalePrice'].median().plot.bar()
    
    plt.title(features)
    plt.show()
```

There are features which have Nan values, yet the SalePrice is high. Hence there is a relationship between the missing values in independent variable and the output dependent variable.

Hence, the missing values need to be handled well which we will do during feature engineering.

#### Temporal Variables

Getting the temporal features...

```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

Only the unique values of the features is shown.

Checking the relationship of 'YearSold' and 'SalePrice'

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel("Year Sold")
plt.ylabel('Median House Price')
plt.title("House Price vs Year Sold")
```

House Price started dipping before the 2008 Housing crisis in USA which started a worldwide recession

```python
year_feature
```

#### As the age of the house increases, is the price increasing?

```python
data = dataset.copy()
for feature in year_feature:
    if feature != 'YrSold':
        data[feature] = data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

Observation: New houses are more costly than the older ones!! Looks like a Pareto or Power Law distribution.

#### Numerical variables are of 2 types:-
    
    1. Discrete
    
    2. Continuous

```python
discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique())<=25]
```

25 is taken as general expert opinion on the number of unique values below which a feature tends to be discrete.

```python
len(discrete_feature)
```

There are 18 discrete features in the dataset.

```python
dataset[discrete_feature].head()
```

##### Relationship between discrete variables and sales price

```python
data = dataset.copy()

for feature in discrete_feature:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(feature)
    plt.show()
```

##### Relationship between Continuous variable and sale price

```python
continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature+year_feature]
```

```python
len(continuous_feature)
```

Using histograms to understand the distribution btw continuous feature and sale price...

```python
data = dataset.copy()

for feature in continuous_feature:
    data.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel(feature)
    plt.ylabel('Sales Price')
    plt.title(feature)
    plt.show()
```

```python
for feature in continuous_feature:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.title(f'count of {feature}')
    plt.show()
```

### EDA : Part 2

Using logarithmic transformation...

```python
data = dataset.copy()
```

```python
for feature in continuous_feature:
    if 0 in data[feature].unique():        # log 0 is undefined
        pass
    else:
        data[feature]=np.log(data[feature])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sales Price')
        plt.title(feature)
        plt.show()
```

### Checking Outliers

```python
data = dataset.copy()

for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
